## Regrouping features

### Values not referenced

```python
data['workclass'] = data['workclass'].replace({'?': 'Not referenced'})

data['native-country'] = data['native-country'].replace({'?': 'Not referenced'})

data['occupation'] = data['occupation'].replace({'?': 'Not referenced'})
```

### Workclass

```python
# Regroup 'Without-pay' and 'Never-worked' to 'No revenu'
data['workclass'] = data['workclass'].replace({'Without-pay': 'No revenu', 'Never-worked': 'No revenu'})

# Regrouper 'Self-emp-not-inc' and 'Self-emp-inc' to 'Self-emp'
data['workclass'] = data['workclass'].replace({'Self-emp-not-inc': 'Self-emp', 'Self-emp-inc': 'Self-emp'})
```

## Modelisation - KNN

```python
X_train = X_train_encoded
X_test = X_test_encoded
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Train the SVM model
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Make predictions
y_pred = svm_model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = [
    {'kernel': ['linear'], 'C': [0.1, 1, 10, 100]},
    {'kernel': ['poly'], 'degree': [2, 3, 4], 'C': [0.1, 1, 10, 100]},
    {'kernel': ['rbf'], 'gamma': [0.1, 0.01, 0.001], 'C': [0.1, 1, 10, 100]},
    {'kernel': ['sigmoid'], 'C': [0.1, 1, 10, 100]}
]

# Initialize Grid Search
grid_search = GridSearchCV(SVC(), param_grid, refit=True, verbose=1, cv=3, n_jobs=-1)

# Perform Grid Search
grid_search.fit(X_train, y_train)

class ProgressPrinter:
    def __init__(self, search, verbose=True):
        self.search = search
        self.verbose = verbose
        
    def print_progress(self):
        if self.verbose:
            means = self.search.cv_results_['mean_test_score']
            stds = self.search.cv_results_['std_test_score']
            params = self.search.cv_results_['params']
            for mean, std, params in zip(means, stds, params):
                print(f"Mean accuracy: {mean:.4f} (±{std:.4f}) with params: {params}")
                print(f"Time taken: {self.search.refit_time_:.2f} seconds")
                print()
                
progress_printer = ProgressPrinter(grid_search)

# Get the best parameters
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

# Train the best model
best_svm_model = grid_search.best_estimator_

# Make predictions
y_pred_best = best_svm_model.predict(X_test)

# Evaluate the fine-tuned model
print(classification_report(y_test, y_pred_best))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the results into a DataFrame
results = pd.DataFrame(grid_search.cv_results_)

# Pivot the DataFrame to create a matrix for heatmap
pivot_table = results.pivot_table(index='param_C', columns='param_gamma', values='mean_test_score')

# Plot the heatmap for RBF kernel
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt=".3f")
plt.title('Grid Search Scores for SVM with RBF Kernel')
plt.xlabel('Gamma')
plt.ylabel('C')
plt.show()

# For linear kernel, we can visualize the results in a different way since gamma is not a parameter for linear kernel
linear_kernel_results = results[results['param_kernel'] == 'linear']

# Plotting C values against mean test score for linear kernel
plt.figure(figsize=(10, 6))
sns.lineplot(data=linear_kernel_results, x='param_C', y='mean_test_score')
plt.xscale('log')
plt.title('Grid Search Scores for SVM with Linear Kernel')
plt.xlabel('C')
plt.ylabel('Mean Test Score')
plt.show()

def visualize_svm(grid_search_):
    # Extract cv_results_ and convert to DataFrame
    results = pd.DataFrame(grid_search_.cv_results_)

    # Visualize mean test scores for different metrics and weights
    plt.figure(figsize=(14, 8))
    sns.lineplot(
        data=results,
        x='param_n_neighbors',
        y='mean_test_score',
        hue='param_metric',
        style='param_weights',
        markers=True,
        dashes=False
    )
    plt.title('Mean Test Scores by n_neighbors, Metric, and Weights')
    plt.xlabel('Number of Neighbors (n_neighbors)')
    plt.ylabel('Mean Test Score')
    plt.legend(title='Metric and Weights')
    plt.show()

    # Create a combined column for metric and weight
    results['metric_weight'] = results['param_metric'].astype(str) + "_" + results['param_weights'].astype(str)

    # Create a pivot table to visualize results as a heatmap
    pivot_table = results.pivot_table(
        values='mean_test_score',
        index='param_n_neighbors',
        columns='metric_weight',
        aggfunc='mean'
    )

    # Plot heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_table, annot=True, cmap='viridis')
    plt.title('Heatmap of Mean Test Scores')
    plt.xlabel('Metric and Weights')
    plt.ylabel('Number of Neighbors (n_neighbors)')
    plt.show()
```
